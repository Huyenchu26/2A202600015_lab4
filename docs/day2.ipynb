```python
import pandas as pd

## Đọc dữ liệu
duLieu = pd.read_csv("users_7_7_2026 2_49_38 AM.csv")
duLieu.head()

## Kiểm tra dữ liệu
duLieu.info()
duLieu.describe(include='all')

## Câu 1
soLuongNguoiDung = len(duLieu)## Sử dụng len(duLieu) để đếm tổng số dòng trong dataset. 
print("Số lượng người dùng:", soLuongNguoiDung)

## Câu 2
usage = (duLieu["Usage location"].value_counts(dropna = False).sort_values(ascending = False))## Sử dụng value_counts() để đếm số lượng người dùng theo từng quốc gia, dropna = False để giữ lại các giá trị NaN. Sau đó sử dụng sort_values() để sắp xếp theo thứ tự giảm dần.
print("Số lượng người dùng theo quốc gia:\n", usage)

## Câu 3
top_department = (duLieu.groupby("Department").agg(Total = ("Department", "count")).sort_values(by = "Total", ascending = False).head(5))## Sử dụng groupby() để nhóm dữ liệu theo cột "Department", sau đó sử dụng agg() để tính tổng số lượng người dùng trong từng phòng ban. Cuối cùng, sử dụng sort_values() để sắp xếp theo thứ tự giảm dần.
print("Top 5 phòng ban có nhiều người dùng nhất:\n", top_department)

## Câu 4
password_true = duLieu.loc[duLieu["Password never expires"] == True]## Sử dụng loc[] để lọc dữ liệu, chỉ lấy những người dùng có cột "Password never expires" là True.
soLuongPasswordTrue = len(password_true)## Sử dụng len() để đếm số lượng người dùng có cột "Password never expires" là True.
percentPasswordTrue = (soLuongPasswordTrue / soLuongNguoiDung) * 100## Tính phần trăm người dùng có cột "Password never expires" là True.
print("Số lượng người dùng có mật khẩu không bao giờ hết hạn:", soLuongPasswordTrue)
print("Phần trăm người dùng có mật khẩu không bao giờ hết hạn: {:.2f}%".format(percentPasswordTrue))

## Câu 5
duLieu["When created"] = pd.to_datetime(duLieu["When created"])## Sử dụng to_datetime() để chuyển đổi cột "When created" sang định dạng datetime.
user_by_year = duLieu.groupby(duLieu["When created"].dt.year).size().sort_index()## Sử dụng groupby() để nhóm dữ liệu theo năm tạo tài khoản, sau đó sử dụng size() để đếm số lượng người dùng trong từng năm. Cuối cùng, sử dụng sort_index() để sắp xếp theo thứ tự tăng dần của năm.
print("Số lượng người dùng theo năm tạo tài khoản:\n", user_by_year)
```
